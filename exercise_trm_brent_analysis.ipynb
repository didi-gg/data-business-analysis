### Introducción al Dólar TRM

El Tipo de Cambio Representativo del Mercado (TRM) es una tasa de cambio que se usa en Colombia para establecer el valor del dólar estadounidense en pesos colombianos (COP). Esta tasa se determina con base en las transacciones de compra y venta del dólar en el mercado colombiano.

### Importancia del Dólar TRM en el Contexto Empresarial

El TRM es crucial para diversas áreas del negocio, como:

- **Importaciones y Exportaciones**: Las empresas que compran o venden productos en dólares deben tener en cuenta el TRM para calcular costos y precios.
- **Planificación Financiera**: Las fluctuaciones en el TRM pueden afectar los ingresos y gastos de las empresas, especialmente para aquellas con operaciones internacionales.
- **Hedging**: Las empresas pueden usar contratos futuros para protegerse contra la volatilidad del TRM.

En esta clase, analizaremos el histórico del Dólar TRM para entender mejor cómo ha cambiado en el tiempo y qué factores podrían influir en estos cambios.

### Objetivo del Análisis

El objetivo es examinar el histórico del TRM para identificar patrones, tendencias y posibles anomalías. Esto ayudará a las empresas a tomar decisiones informadas y a prever posibles impactos en sus operaciones financieras.

### Transformaciones Mínimas

1. **Conversión de Tipos de Datos**: Convertiremos las columnas `Fecha`, `Último`, `Apertura`, `Máximo`, `Mínimo`, `Vol.` y `% var.` a tipos de datos adecuados.
2. **Manejo de Datos Faltantes**: Imputaremos los valores faltantes en la columna `Vol.` con la mediana de la columna.
3. **Limpieza de Datos**: Eliminaremos caracteres no numéricos y transformaremos las comas en puntos decimales en las columnas numéricas.


````python
brent_df['Fecha'] = pd.to_datetime(brent_df['Fecha'], format='%d.%m.%Y')
````

````python
brent_df['Fecha'].head()
````

````python
# Convertimos las columnas numéricas de 'object' a 'float'
cols_to_convert = ['Último', 'Apertura', 'Máximo', 'Mínimo', 'Vol.', '% var.']

for col in cols_to_convert:
    brent_df[col] = brent_df[col].str.replace('.', '', regex=False)  # Removemos puntos de miles
    brent_df[col] = brent_df[col].str.replace(',', '.', regex=False)  # Reemplazamos comas por puntos decimales
    brent_df[col] = brent_df[col].str.replace('K', 'e3', regex=False)  # Convertimos 'K' a notación científica
    brent_df[col] = brent_df[col].str.replace('%', '', regex=False)  # Removemos el símbolo de porcentaje
    brent_df[col] = pd.to_numeric(brent_df[col], errors='coerce')
````

2. Transformación de los datos

````python
# Función para limpiar y estandarizar nombres de columnas
def clean_column_names(df):
    # Crear una tabla de traducción para eliminar tildes
    translation_table = str.maketrans(
        'áéíóúüñÁÉÍÓÚÜÑ',
        'aeiouunAEIOUUN'
    )

    # Limpiar y estandarizar nombres de columnas
    df.columns = [
        # Traducir tildes
        col.translate(translation_table)
        # Reemplazar caracteres especiales por guiones bajos
        .replace(' ', '_')
        .replace('.', '')
        .replace('%', 'percent')
        .replace(',', '')
        # Convertir a minúsculas
        .lower()
        for col in df.columns
    ]
    return df
````

````python
# Aplicamos la limpieza de nombres de columnas
trm_transformed_df = clean_column_names(trm_df)
brent_transformed_df = clean_column_names(brent_df)
````

````python
brent_transformed_df.head()
````

````python
trm_transformed_df.rename(columns={'vigenciadesde': 'fecha'}, inplace=True)
trm_transformed_df.rename(columns={'valor': 'valor_trm'}, inplace=True)
brent_transformed_df.rename(columns={'ultimo': 'valor_brent'}, inplace=True)
````

````python
trm_transformed_df.columns
````

````python
trm_transformed_df.set_index('fecha', inplace=True)
brent_transformed_df.set_index('fecha', inplace=True)
````

````python
# Verificamos si hay duplicados en los índices y los eliminamos si existen
trm_transformed_df = trm_transformed_df[~trm_transformed_df.index.duplicated(keep='first')]
brent_transformed_df = brent_transformed_df[~brent_transformed_df.index.duplicated(keep='first')]
````

````python
common_dates = trm_transformed_df.index.intersection(brent_transformed_df.index)
````

````python
# Reindexar ambos DataFrames para que tengan solo las fechas comunes
trm_transformed_df = trm_transformed_df.loc[common_dates]
brent_transformed_df = brent_transformed_df.loc[common_dates]
````

````python
# Reindexar ambos DataFrames para que tengan solo las fechas comunes
trm_transformed_df = trm_transformed_df.loc[common_dates]
brent_transformed_df = brent_transformed_df.loc[common_dates]

# Combinamos los datasets en función de la fecha, manteniendo solo las fechas comunes
combined_df = pd.merge(
    brent_transformed_df,
    trm_transformed_df,
    left_index=True,
    right_index=True,
    suffixes=('_brent', '_trm')
)

# Verificamos el DataFrame combinado
print("\nDatos combinados:")
combined_df.head()
````

````python
# Evaluar correlación entre 'valor' y 'valor_brent'
correlation = combined_df[['valor_brent', 'valor_trm']].corr()
print("\nCorrelación entre 'valor_brent' y 'valor_trm':")
print(correlation)
````

````python
# Visualizar la correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlación entre TRM y Brent')
plt.show()
````

````python
# Identificación y eliminación de outliers usando el método IQR (Interquartile Range)
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
    return df

# Eliminar outliers
combined_df_cleaned = remove_outliers(combined_df, ['valor_brent', 'valor_trm'])

# Verificamos el DataFrame limpio
print("\nDatos después de eliminar outliers:")
combined_df_cleaned.head()
````

````python
from sklearn.preprocessing import StandardScaler

# Normalización de datos
scaler = StandardScaler()

# Aplicar la normalización a los datos combinados
combined_df_cleaned[['valor_brent', 'valor_trm']] = scaler.fit_transform(combined_df_cleaned[['valor_brent', 'valor_trm']])

# Verificamos los datos normalizados
print("\nDatos normalizados:")
combined_df_cleaned.head()
````

````python
import matplotlib.pyplot as plt

# Graficar los datos de valor_brent y valor_trm a lo largo del tiempo
plt.figure(figsize=(14, 7))

# Línea para 'valor_brent'
plt.plot(combined_df_cleaned.index, combined_df_cleaned['valor_brent'],
         label='Valor Brent', color='blue', linestyle='-', linewidth=2)

# Línea para 'valor_trm'
plt.plot(combined_df_cleaned.index, combined_df_cleaned['valor_trm'],
         label='Valor TRM', color='red', linestyle='-', linewidth=2)

# Añadir títulos y etiquetas
plt.title('Comparación de Valor Brent y Valor TRM a lo largo del tiempo', fontsize=16)
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Valor Normalizado', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Mejorar la visualización
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
